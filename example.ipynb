```python
from src import encrypt as enc
from src import decrypt as dec
from src import genKey as gk
```

```python
img = cv2.imread('image/bishil06.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.imshow(img_rgb)
plt.show()
```

```python
keyList = gk.gen_key(img, 1)
print(keyList)
```

```python
import time
start = time.time()
enc_img = enc.encrypt(img_rgb, keyList)
end = time.time()
print('enc time :', end - start, 'second')
```

```python
plt.imshow(enc_img)
plt.show()
```

```python
start = time.time()
dec_img = dec.decrypt(enc_img, keyList)
end = time.time()
print('dec time :', end - start, 'second')
```

```python
plt.imshow(dec_img)
plt.show()
```
